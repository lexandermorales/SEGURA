```python
#import libraries
import pandas as pd
import nltk 
import spacy
import re
import numpy as np
from nltk.stem.porter import PorterStemmer
from nltk.stem.snowball import SnowballStemmer
sn = SnowballStemmer(language='spanish')
ps = PorterStemmer()
nlp = spacy.load("es_core_news_sm")
#nltk.download('stopwords')
from nltk.corpus import stopwords

# In[222]:


#import dataset
dataset = pd.read_csv("dataset.tsv")
dataset = pd.DataFrame(dataset)
text_sentiment = dataset[['Texto', 'Sentimiento']]
text_sentiment.dropna(inplace=True)
text_sentiment.reset_index(drop=True,  inplace=True)
text_sentiment.reset_index(drop=True)
#Change a,b,c -> 1,0.5,0
```

```python
from nltk.stem.porter import PorterStemmer
from nltk.stem.snowball import SnowballStemmer
sn = SnowballStemmer(language='spanish')
ps = PorterStemmer()
nlp = spacy.load("es_core_news_sm")
#nltk.download('stopwords')
from nltk.corpus import stopwords
```

```python
corpus = []
def preprocessing():
    for i in range(0,len(text_sentiment)):
        pat = r'[^a-zA-z칌침x치칠칤칩칰]'
        review = re.sub(pat, ' ', text_sentiment["Texto"][i])
        review = review.split()
        review = ' '.join(review)
        doc = nlp(review)
        review = []
        for token in doc:
            words_tokenize = token.lemma_
            words_tokenize = words_tokenize.lower()
            words_tokenize = ps.stem(words_tokenize)
            words_tokenize = sn.stem(words_tokenize)
            review.append(words_tokenize)
        review = [ps.stem(word) for word in review if not word in set(stopwords.words('spanish'))]
        review = ' '.join(review)
        new = []
        new.append(review)
        corpus.append(new)
        
   
    
```

```python
preprocessing()
```

```python
corpus = [item for val in corpus for item in val]
```
